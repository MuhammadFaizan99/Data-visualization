# Boxplot

```python
# Always between 1 continuous and 1 categorical variable
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
boat=sns.load_dataset("titanic")
sns.boxplot(x="sex",y="age",data=boat)
plt.title("BoxPlot of titanic")
```

## Boxplot with different dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
tip=sns.load_dataset("tips")
sns.boxplot(x="time",y="total_bill",data=tip)
plt.title("BoxPlot of Tips")
```

### Saturation

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
tip=sns.load_dataset("tips")
sns.boxplot(x="time",y="total_bill",data=tip,saturation=0.4)
plt.title("BoxPlot of Tips")
```

### Box  Plot for 1 variable

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x=tip['total_bill'])
```

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",hue="smoker",data=tip)
```

### Palette

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",hue="smoker",data=tip,palette="Set2")
```

### Dodge

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",hue="smoker",data=tip,palette="Set2",dodge=True)
```

### Color

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",hue="smoker",data=tip,color="yellow")
```

```python
import seaborn as sns
sns.set(style="whitegrid")
boat=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",data=boat,showmeans=True)
```

```python
import seaborn as sns
sns.set(style="whitegrid")
boat=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",data=boat,showmeans=True,meanprops={"marker":"*","markersize":"12"
                                                                     ,"markeredgecolor":"red"})
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
boat=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",data=boat,showmeans=True,meanprops={"marker":"*","markersize":"32"
                                                                     ,"markeredgecolor":"red"})
plt.xlabel("No of survived in incident",size=13)
plt.ylabel("Age of survival people",size=15)
plt.title("Boxplot",size=12,weight='bold')
```
