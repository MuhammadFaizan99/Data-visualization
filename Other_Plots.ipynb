# Line plot with multifacets

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dot=sns.load_dataset("dots")
p=sns.color_palette('rocket_r')
sns.relplot(data=dot,x="time",y="firing_rate",hue="coherence",size="choice",col="align",kind="line",
            size_order=["T1","T2"],palette=p,height=5,aspect=.75,facet_kws=dict(sharex=False))
```
