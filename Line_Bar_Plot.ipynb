```python
# Line Plot:Always between two numeric veribles
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
flower=sns.load_dataset("iris")
sns.lineplot(x="sepal_length",y="petal_width",data=flower)
plt.title("Line Plot of Flower")
plt.show()
```

### Adding Limits

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
flower=sns.load_dataset("iris")
sns.lineplot(x="sepal_length",y="petal_width",data=flower)
plt.title("Line Plot of Flower")
plt.xlim(2)
plt.ylim(0.2)
plt.show()
```

### Size of Figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
flower=sns.load_dataset("iris")
plt.figure(figsize=(4,2))
sns.lineplot(x="sepal_length",y="petal_width",data=flower)
plt.title("Line Plot of Flower")
sns.set_style("darkgrid")
sns.set_style(style=None,rc=None)
plt.show()
```

# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
sns.barplot(x="sex",y="fare",hue="class",data=boat)
plt.title("Bar Plot of Titanic")
plt.show()
```

### Order Changing

```python
import seaborn as sns
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
sns.barplot(x="sex",y="fare",hue="class",data=boat,order=["female","male"])
plt.title("Bar Plot of Titanic")
plt.show()
```

### Removing Error Bars

```python
import seaborn as sns
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
sns.barplot(x="sex",y="fare",hue="class",data=boat,order=["female","male"],color="green",ci=None)
plt.title("Bar Plot of Titanic")
plt.show()
```

### Estimator

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
sns.barplot(x="sex",y="fare",hue="class",data=boat,order=["female","male"],color="green"
            ,palette="YlOrBr",estimator=mean)
plt.title("Bar Plot of Titanic")
plt.show()
```

### Saturation

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
sns.barplot(x="sex",y="fare",hue="class",data=boat,order=["female","male"],color="green"
            ,palette="YlOrBr",estimator=mean,saturation=0.5)
plt.title("Bar Plot of Titanic")
plt.show()
```
